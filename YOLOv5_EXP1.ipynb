```python
import cv2
import os
import torch
import matplotlib.pyplot as plt
from deepface import DeepFace
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings('ignore')
%matplotlib inline
```

```python
model = torch.hub.load('./yolov5/', 'custom', path='./yolov5/best1.pt', source='local')
```

```python
os.listdir("Face_data")
```

```python
cap=cv2.VideoCapture(0)
while True:
    check,frame=cap.read()
    if not check:
        break
    results=model(frame)
    results=results.pandas()
    for i in range(len(results.xyxyn[0])):
        ll=(int)(results.xyxy[0].iloc[i].xmin)
        rl=(int)(results.xyxy[0].iloc[i].ymin)
        lr=(int)(results.xyxy[0].iloc[i].xmax)
        rr=(int)(results.xyxy[0].iloc[i].ymax)
        conf=results.xyxyn[0].iloc[i].confidence
        l_name = results.xyxyn[0].iloc[i]['name']
        label = results.xywh[0].iloc[i]['class']
        
        if l_name == 'face':
            
            try:
#                 img = frame[ll:lr,rl:rr]
                res = DeepFace.find(img_path = frame,db_path = './Face_Data/',model_name='Facenet')

#                 print(res)
                if(len(res)>0):
                    
                    l_id = res['identity'][0].split('/')[3].split('.')[0]
                    
                    cv2.rectangle(frame,(ll,rl),(lr,rr),(255,0,0),2)
                    cv2.putText(frame,l_id,(ll, rl-10),cv2.FONT_HERSHEY_SIMPLEX,1,(255,255,255),2)
                else:
                    cv2.rectangle(frame,(ll,rl),(lr,rr),(255,0,0),2)
                    cv2.putText(frame,l_name,(ll, rl-10),cv2.FONT_HERSHEY_SIMPLEX,1,(255,255,255),2) 
            except:
                cv2.rectangle(frame,(ll,rl),(lr,rr),(255,0,0),2)
                cv2.putText(frame,l_name,(ll, rl-10),cv2.FONT_HERSHEY_SIMPLEX,1,(255,255,255),2) 
        else:
            cv2.rectangle(frame,(ll,rl),(lr,rr),(255,255,0),2)
            cv2.putText(frame,l_name,(ll, rl-10),cv2.FONT_HERSHEY_SIMPLEX,1,(255,255,0),2)
    cv2.imshow('Video',frame)
    
    if(cv2.waitKey(1) & 0XFF==27):
        break
cap.release()
cv2.destroyAllWindows()
```
